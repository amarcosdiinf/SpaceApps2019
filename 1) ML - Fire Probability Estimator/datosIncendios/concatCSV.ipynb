```python
import csv
import pandas as pd
import os
import sys
```

```python
ls = ['MODIS C6 - 2015',
      'MODIS C6 - 2016',
      'MODIS C6 - 2017',
      'MODIS C6 - 2018',
      'MODIS C6 - 2019']
ls2 = ['VIIRS - 2015',
       'VIIRS - 2016',
       'VIIRS - 2017',
       'VIIRS - 2018',
       'VIIRS - 2019']
ls
```

```python
files = []
for folder in ls:
    path = './' + folder

    # r=root, d=directories, f = files
    for r, d, f in os.walk(path):
        for file in f:
            if '.csv' in file:
                files.append(os.path.join(r, file))
```

```python
files
```

```python
combined_csv = pd.concat([pd.read_csv(f) for f in files ])
#export to csv
combined_csv.to_csv( "datosModis.csv", index=False, encoding='utf-8-sig')
```
