```python
import csv
import pandas as pd
import os
import sys
import http.client
import ssl
import json
import datetime
import requests
import time
from datetime import timedelta, date
```

```python
def getData(latitud, longitud):
    url = "https://us1.locationiq.com/v1/reverse.php"

    data = {
        'key': 'b3483c524735f8',
        'lat': str(latitud),
        'lon': str(longitud),
        'format': 'json'
    }

    response = requests.get(url, params=data)
    try:
        provincia = response.json()['address']['county']
        
    except:
        provincia = 'NaN'
    
    if provincia=='NaN':
        try:
            provincia = response.json()['address']['state']
            if provincia == 'Biscay':
                provincia = response.json()['address']['city']
        except:
            provincia = 'NaN'
    
    #time.sleep(2)
    return provincia
```

```python
dfModis = pd.read_csv('./MODIS C6 - 2018/fire_archive_M6_81164.csv')
dfViirs = pd.read_csv('./VIIRS - 2018/fire_archive_V1_81165.csv')
```

```python
df2018 = pd.concat([dfModis, dfViirs], axis = 0)
df2018.head(5)
```

```python
df2018=df2018[['acq_date', 'latitude', 'longitude']]
df2018.head(5)
```

```python
df2018['provincias']=df2018.apply(lambda x : getData(x.latitude, x.longitude), axis=1)
```
