```python
import csv
import pandas as pd
import os
import sys
```

```python
path = '.'

files=[]
# r=root, d=directories, f = files
for r, d, f in os.walk(path):
    for file in f:
        if '.csv' in file:
            files.append(os.path.join(r, file))
```
